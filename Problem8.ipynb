## 169. Majority Element

image.png

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        candidate = None  # Biến để lưu phần tử ứng cử viên
        count = 0  # Biến để đếm số lần xuất hiện của ứng cử viên

        # Duyệt qua từng số trong mảng nums
        for num in nums:
            if count == 0:
                candidate = num  # Đặt số hiện tại là ứng cử viên mới

            if num == candidate:
                count += 1  # Nếu số hiện tại là ứng cử viên, tăng đếm
            else:
                count -= 1  # Nếu không, giảm đếm

        return candidate  # Trả về ứng cử viên, đây là phần tử chiếm ưu thế
```

- Biến candidate: Dùng để lưu phần tử ứng cử viên có thể là phần tử chiếm ưu thế.
- Biến count: Dùng để đếm số lần xuất hiện của phần tử ứng cử viên.
- Khi count bằng 0, đặt phần tử hiện tại là ứng cử viên mới.
Nếu phần tử hiện tại bằng ứng cử viên, tăng count.
Nếu không, giảm count.
- Ứng cử viên cuối cùng sẽ là phần tử chiếm ưu thế, vì theo định nghĩa, nó xuất hiện nhiều hơn n/2 lần.
