## 217. Contains Duplicate

image.png

```python
class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        seen = []  # Tạo một danh sách rỗng để theo dõi các số đã thấy

        # Lặp qua từng số trong mảng
        for num in nums:
            # Kiểm tra xem số này đã có trong danh sách chưa
            if num in seen:
                return True  # Nếu có, thì chúng ta đã tìm thấy số trùng!
            seen.append(num)  # Nếu chưa, thêm số vào danh sách đã thấy

        return False  # Nếu hoàn tất vòng lặp mà không tìm thấy số trùng nào
```

- Danh sách seen: Đây là nơi chúng ta sẽ lưu các số đã gặp trong quá trình lặp qua mảng nums.

- Kiểm tra trùng lặp: Nếu số hiện tại (num) đã tồn tại trong danh sách seen, điều đó có nghĩa là nó đã xuất hiện trước đó, và chúng ta trả về True.
Nếu số chưa xuất hiện, chúng ta thêm nó vào danh sách seen.
- Nếu sau khi kiểm tra tất cả các số mà không tìm thấy số nào trùng, chúng ta trả về False.
