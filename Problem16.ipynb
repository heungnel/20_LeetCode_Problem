## 1046. Last Stone Weight

image.png

```python
class Solution(object):
    def lastStoneWeight(self, stones):
        """
        :type stones: List[int]
        :rtype: int
        """
        # Sắp xếp mảng stones để dễ dàng tìm hai viên đá nặng nhất
        stones.sort()

        # Trong khi còn ít nhất 2 viên đá
        while len(stones) > 1:
            # Lấy hai viên đá nặng nhất
            stone1 = stones.pop()  # Viên đá nặng nhất
            stone2 = stones.pop()  # Viên đá nặng thứ hai

            # Thực hiện đập đá
            if stone1 != stone2:
                # Nếu hai viên đá khác trọng lượng, đưa sự chênh lệch trở lại mảng
                stones.append(stone1 - stone2)
                stones.sort()  # Sắp xếp lại mảng sau khi thêm viên đá mới

        # Trả về trọng lượng viên đá cuối cùng hoặc 0 nếu không còn viên đá nào
        return stones[0] if stones else 0
```

- stones.sort() sắp xếp mảng stones theo thứ tự tăng dần. Điều này giúp dễ dàng lấy hai viên đá nặng nhất khi cần thiết.

- while len(stones) > 1: đảm bảo rằng còn ít nhất hai viên đá trong mảng để có thể thực hiện smash.

- Lấy hai viên đá nặng nhất:

stone1 = stones.pop() lấy viên đá nặng nhất (cuối cùng trong danh sách sau khi sắp xếp).

stone2 = stones.pop() lấy viên đá nặng thứ hai.
Thực hiện smash:

- Nếu hai viên đá có trọng lượng khác nhau (if stone1 != stone2:), ta tính trọng lượng viên đá mới bằng cách lấy hiệu số và đưa nó vào mảng stones:
stones.append(stone1 - stone2) thêm viên đá mới (sự chênh lệch) vào mảng.
stones.sort() sắp xếp lại mảng để chuẩn bị cho lần smash tiếp theo.

- Cuối cùng, nếu còn viên đá trong mảng, ta trả về trọng lượng của nó (return stones[0]). Nếu không còn viên đá nào, ta trả về 0.
