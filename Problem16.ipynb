## 1046. Last Stone Weight

image.png

```python
class Solution(object):
    def lastStoneWeight(self, stones):
        """
        :type stones: List[int]
        :rtype: int
        """
        # Sp xp mng stones  d dng tm hai vin  nng nht
        stones.sort()

        # Trong khi cn t nht 2 vin 
        while len(stones) > 1:
            # Ly hai vin  nng nht
            stone1 = stones.pop()  # Vin  nng nht
            stone2 = stones.pop()  # Vin  nng th hai

            # Thc hin p 
            if stone1 != stone2:
                # Nu hai vin  khc trng lng, a s chnh lch tr li mng
                stones.append(stone1 - stone2)
                stones.sort()  # Sp xp li mng sau khi thm vin  mi

        # Tr v trng lng vin  cui cng hoc 0 nu khng cn vin  no
        return stones[0] if stones else 0
```

- stones.sort() sp xp mng stones theo th t tng dn. iu ny gip d dng ly hai vin  nng nht khi cn thit.

- while len(stones) > 1: m bo rng cn t nht hai vin  trong mng  c th thc hin smash.

- Ly hai vin  nng nht:

stone1 = stones.pop() ly vin  nng nht (cui cng trong danh sch sau khi sp xp).

stone2 = stones.pop() ly vin  nng th hai.
Thc hin smash:

- Nu hai vin  c trng lng khc nhau (if stone1 != stone2:), ta tnh trng lng vin  mi bng cch ly hiu s v a n vo mng stones:
stones.append(stone1 - stone2) thm vin  mi (s chnh lch) vo mng.
stones.sort() sp xp li mng  chun b cho ln smash tip theo.

- Cui cng, nu cn vin  trong mng, ta tr v trng lng ca n (return stones[0]). Nu khng cn vin  no, ta tr v 0.
