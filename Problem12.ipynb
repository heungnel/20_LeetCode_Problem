## 258. Add Digits

image.png

```python
class Solution(object):
    def addDigits(self, num):
        """
        :type num: int
        :rtype: int
        """
        while num >= 10:  # Lặp lại cho đến khi num còn là một chữ số
            total = 0
            while num > 0:  # Tính tổng các chữ số của num
                total += num % 10  # Cộng chữ số cuối cùng vào tổng
                num //= 10  # Loại bỏ chữ số cuối cùng của num
            num = total  # Cập nhật giá trị num bằng tổng các chữ số
        return num  # Trả về kết quả khi num còn lại một chữ số
```

- Kiểm tra số có phải là chữ số đơn không: Nếu num >= 10, nghĩa là num có từ hai chữ số trở lên, ta sẽ tiếp tục vòng lặp để tính tổng các chữ số.

-  Sử dụng vòng lặp while num > 0, ta lần lượt lấy từng chữ số cuối của num (bằng cách lấy phần dư của num chia cho 10) và cộng vào biến total.

- Sau khi tính xong tổng các chữ số, ta cập nhật giá trị của num bằng total.

- Khi num nhỏ hơn 10 (chỉ còn 1 chữ số), vòng lặp kết thúc và ta trả về kết quả.
