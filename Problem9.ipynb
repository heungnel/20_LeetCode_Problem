## 202. Happy Number

image.png

```python
class Solution(object):
    def isHappy(self, n):
        """
        :type n: int
        :rtype: bool
        """
        # Hàm tính tổng bình phương của các chữ số
        def sum_of_squares(num):
            total = 0
            while num > 0:
                digit = num % 10  # Lấy chữ số cuối cùng
                total += digit ** 2  # Bình phương chữ số và cộng vào tổng
                num //= 10  # Loại bỏ chữ số cuối cùng
            return total

        seen_numbers = []  # Danh sách để theo dõi các số đã thấy

        while n != 1:  # Tiếp tục cho đến khi n bằng 1
            if n in seen_numbers:  # Kiểm tra xem số này đã được thấy chưa
                return False  # Nếu đã thấy -> không phải là số hạnh phúc
            seen_numbers.append(n)  # Thêm số hiện tại vào danh sách
            n = sum_of_squares(n)  # Cập nhật n thành tổng bình phương của các chữ số

        return True  # Nếu đạt tới 1 -> thì là số hạnh phúc
```

- Hàm sum_of_squares(num): Tính tổng bình phương của các chữ số của số num.

- Sử dụng vòng lặp để lấy từng chữ số, tính bình phương và cộng vào total.
Loại bỏ chữ số cuối cùng bằng phép chia nguyên.
Danh sách seen_numbers: Lưu trữ các số đã gặp để phát hiện vòng lặp. Nếu một số xuất hiện lại, điều đó có nghĩa là chúng ta đang trong một chu kỳ và không thể đạt đến 1.

- Tiếp tục cho đến khi n bằng 1. Kiểm tra nếu n đã xuất hiện trong seen_numbers. Nếu có, trả về False vì không phải là số hạnh phúc. Thêm n vào danh sách và cập nhật n thành tổng bình phương của các chữ số.
- Nếu đạt được n = 1, hàm trả về True, chứng tỏ số là số hạnh phúc.
