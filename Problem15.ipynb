## 448. Find All Numbers Disappeared in an Array

image.png

```python
class Solution(object):
    def findDisappearedNumbers(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        # Tính độ dài của mảng nums
        n = len(nums)

        # Chuyển mảng nums thành một tập hợp để dễ dàng tìm kiếm các giá trị
        a = set(nums)

        # Tạo một danh sách rỗng để lưu các giá trị bị thiếu
        missing_value = []

        # Duyệt qua tất cả các số từ 1 đến n
        for i in range(1, n + 1):  # Duyệt từ 1 đến n (không tính 0)
            # Nếu số i không có trong tập hợp a, nó bị thiếu trong mảng
            if i not in a:
                # Thêm số bị thiếu vào danh sách missing_value
                missing_value.append(i)

        # Trả về danh sách các số bị thiếu
        return missing_value
```

- Tính độ dài của mảng n = len(nums) tính độ dài của mảng nums. Điều này giúp xác định phạm vi từ 1 đến n mà chúng ta cần kiểm tra.

- Chuyển đổi thành tập hợp a = set(nums) chuyển đổi mảng nums thành một tập hợp.

- Tạo danh sách rỗng missing_value = [] tạo ra một danh sách rỗng để lưu trữ các số bị thiếu.

- Duyệt qua phạm vi [1, n] for i in range(1, n + 1) duyệt qua tất cả các số từ 1 đến n. Nếu số đó không tồn tại trong tập hợp a, nó là số bị thiếu.

- Nếu một số không tồn tại trong tập hợp, nó sẽ được thêm vào danh sách missing_value.

- Sau khi kiểm tra toàn bộ phạm vi, danh sách missing_value chứa các số không xuất hiện trong mảng ban đầu sẽ được trả về.
