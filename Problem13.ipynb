## 263. Ugly Number

image.png

```python
class Solution(object):
    def isUgly(self, n):
        """
        :type n: int
        :rtype: bool
        """
        # Step 1: Handle the special case for non-positive integers
        if n <= 0:
            return False  # Ugly numbers are positive integers

        # Step 2: Divide by 2, 3, and 5
        for prime in [2, 3, 5]:
            while n % prime == 0:  # Check if n can be divided by the prime
                n //= prime  # Keep dividing n by the prime factor

        # Step 3: Check if n is reduced to 1
        return n == 1
```

- Nếu n là số nhỏ hơn hoặc bằng 0, trả về False vì số xấu phải là số dương.
Nếu n == 1, trả về True, vì 1 không có các ước nguyên tố và được coi là một số xấu.
- Sử dụng vòng lặp để chia liên tục n cho 2, 3, hoặc 5, miễn là n chia hết cho các số đó. Nếu không chia hết nữa, chuyển sang số nguyên tố tiếp theo.
- Nếu sau khi chia hết, n còn lại bằng 1, thì đó là một số xấu. Nếu không, n chứa các ước nguyên tố khác ngoài 2, 3, 5, và ta trả về False.
