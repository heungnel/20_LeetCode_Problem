## 1518. Water Bottles

image.png

```python
class Solution(object):
    def numWaterBottles(self, numBottles, numExchange):
        """
        :type numBottles: int
        :type numExchange: int
        :rtype: int
        """
        total_drunk = 0  # Tng s chai nc  ung
        full_bottles = numBottles  # Bt u vi s chai y ban u
        empty_bottles = 0  # Bt u vi khng c chai rng

        while full_bottles > 0:  # Tip tc khi cn chai y
            # Ung tt c cc chai y
            total_drunk += full_bottles  # Cng s chai ung vo tng
            empty_bottles += full_bottles  # Tt c tr thnh chai rng
            full_bottles = 0  #  ung ht cc chai y

            # Kim tra xem c th i chai rng ly chai y khng
            full_bottles = empty_bottles // numExchange  # i chai rng ly chai y
            empty_bottles = empty_bottles % numExchange  # Chai rng cn li

        return total_drunk  # Tr v tng s chai nc  ung
```

- total_drunk: Bin lu tr tng s chai nc  ung, khi to bng 0.
full_bottles: S chai y ban u (numBottles).
empty_bottles: S chai rng ban u, khi to bng 0.

- S dng vng lp while full_bottles > 0:  tip tc cho n khi khng cn chai y no.

- Ung chai y: Cng s chai y vo total_drunk v cp nht s chai rng.
nh du chai  ung: t full_bottles thnh 0 v  ung ht tt c cc chai.

- Tnh s chai y c th nhn c t s chai rng: full_bottles = empty_bottles // numExchange.
Cp nht s chai rng cn li sau khi i: empty_bottles = empty_bottles % numExchange.

- Khi khng cn chai y, vng lp dng li v hm tr v tng s chai nc  ung (total_drunk).
