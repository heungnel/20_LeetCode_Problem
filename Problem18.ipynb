## 1518. Water Bottles

image.png

```python
class Solution(object):
    def numWaterBottles(self, numBottles, numExchange):
        """
        :type numBottles: int
        :type numExchange: int
        :rtype: int
        """
        total_drunk = 0  # Tổng số chai nước đã uống
        full_bottles = numBottles  # Bắt đầu với số chai đầy ban đầu
        empty_bottles = 0  # Bắt đầu với không có chai rỗng

        while full_bottles > 0:  # Tiếp tục khi còn chai đầy
            # Uống tất cả các chai đầy
            total_drunk += full_bottles  # Cộng số chai uống vào tổng
            empty_bottles += full_bottles  # Tất cả trở thành chai rỗng
            full_bottles = 0  # Đã uống hết các chai đầy

            # Kiểm tra xem có thể đổi chai rỗng lấy chai đầy không
            full_bottles = empty_bottles // numExchange  # Đổi chai rỗng lấy chai đầy
            empty_bottles = empty_bottles % numExchange  # Chai rỗng còn lại

        return total_drunk  # Trả về tổng số chai nước đã uống
```

- total_drunk: Biến lưu trữ tổng số chai nước đã uống, khởi tạo bằng 0.
full_bottles: Số chai đầy ban đầu (numBottles).
empty_bottles: Số chai rỗng ban đầu, khởi tạo bằng 0.

- Sử dụng vòng lặp while full_bottles > 0: để tiếp tục cho đến khi không còn chai đầy nào.

- Uống chai đầy: Cộng số chai đầy vào total_drunk và cập nhật số chai rỗng.
Đánh dấu chai đã uống: Đặt full_bottles thành 0 vì đã uống hết tất cả các chai.

- Tính số chai đầy có thể nhận được từ số chai rỗng: full_bottles = empty_bottles // numExchange.
Cập nhật số chai rỗng còn lại sau khi đổi: empty_bottles = empty_bottles % numExchange.

- Khi không còn chai đầy, vòng lặp dừng lại và hàm trả về tổng số chai nước đã uống (total_drunk).
