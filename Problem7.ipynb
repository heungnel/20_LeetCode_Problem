## 136. Single Number

image.png

```python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # Tạo một từ điển để đếm số lần xuất hiện của mỗi số
        count_dict = {}

        # Duyệt qua từng số trong mảng nums
        for num in nums:
            # Nếu số đã có trong từ điển, tăng số đếm của nó
            if num in count_dict:
                count_dict[num] += 1
            else:
                # Nếu chưa có, thêm nó vào từ điển với số đếm là 1
                count_dict[num] = 1

        # Duyệt qua từ điển để tìm số xuất hiện một lần
        for num, count in count_dict.items():
            if count == 1:  # Chúng ta đang tìm số có số đếm là 1
                return num  # Trả về số duy nhất
```

- Tạo từ điển count_dict dùng để đếm số lần xuất hiện của mỗi số trong mảng nums.
- Duyệt qua mảng nums
Nếu số đã có trong từ điển, tăng số đếm.
Nếu chưa có, thêm số vào từ điển với số đếm là 1.
- Duyệt qua từ điển để tìm số có số đếm bằng 1 và trả về số đó.
