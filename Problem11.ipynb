## 242. Valid Anagram

image.png

```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        # nếu độ dài của hai chuỗi khác nhau thì không thể là anagram
        if len(s) != len(t):
            return False

        # Tạo một từ điển để đếm số lần xuất hiện của các ký tự trong chuỗi `s`
        count = {}

        # Đếm các ký tự trong chuỗi `s`
        for char in s:
            if char in count:
                count[char] += 1  # Nếu ký tự đã tồn tại, tăng giá trị đếm lên 1
            else:
                count[char] = 1  # Nếu ký tự chưa tồn tại, khởi tạo giá trị đếm là 1

        # Giảm giá trị đếm dựa trên các ký tự trong chuỗi `t`
        for char in t:
            if char in count:
                count[char] -= 1  # Nếu ký tự tồn tại trong từ điển, giảm giá trị đếm đi 1
            else:
                return False  # Nếu ký tự không tồn tại trong `s`, thì `t` không thể là anagram

        # Kiểm tra xem tất cả các giá trị đếm có bằng 0 hay không
        for value in count.values():
            if value != 0:
                return False  # Nếu có bất kỳ giá trị nào khác 0, thì không phải là anagram

        return True  # Nếu tất cả các giá trị đếm đều bằng 0, thì `t` là anagram của `s`
```

- Nếu hai chuỗi có độ dài khác nhau, chắc chắn rằng chúng không thể là anagram, do đó trả về False ngay lập tức.

- Ta sử dụng một từ điển để đếm số lần xuất hiện của từng ký tự trong chuỗi s.

- Khi duyệt qua chuỗi t, nếu ký tự tồn tại trong từ điển, ta giảm giá trị tương ứng. Nếu xuất hiện ký tự trong t mà không có trong s, ta trả về False.

- Nếu sau khi duyệt qua cả hai chuỗi, tất cả giá trị đếm đều bằng 0, điều này có nghĩa là số lượng ký tự giữa hai chuỗi là giống nhau, do đó chuỗi t là anagram của chuỗi s.
