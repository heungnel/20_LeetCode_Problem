## 121. Best Time to Buy and Sell Stock

image.png

```python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        min_price = prices[0]  # Khởi tạo giá mua tối thiểu bằng giá cổ phiếu đầu tiên
        max_profit = 0  # Khởi tạo lợi nhuận tối đa là 0

        # Duyệt qua các mức giá
        for price in prices:
            # Cập nhật giá mua tối thiểu nếu giá hiện tại thấp hơn
            if price < min_price:
                min_price = price
            # Tính lợi nhuận tiềm năng và cập nhật lợi nhuận tối đa nếu lợi nhuận này cao hơn
            elif price - min_price > max_profit:
                max_profit = price - min_price
        return max_profit  # Trả về lợi nhuận tối đa
```

- min_price được khởi tạo bằng giá cổ phiếu đầu tiên, đây là giá thấp nhất mà chúng ta đã thấy.
max_profit được khởi tạo là 0, đây là lợi nhuận tối đa mà chúng ta sẽ tính toán.
- Sử dụng vòng lặp for để duyệt từng giá cổ phiếu trong danh sách prices:
Nếu giá hiện tại (price) thấp hơn min_price, cập nhật min_price.
Tính lợi nhuận tiềm năng bằng cách lấy chênh lệch giữa giá hiện tại và min_price. Nếu lợi nhuận tiềm năng này lớn hơn max_profit, cập nhật max_profit.
- Sau khi duyệt toàn bộ giá cổ phiếu, hàm trả về max_profit, tức là lợi nhuận tối đa có thể đạt được từ giao dịch.

# Ví dụ:
Với đầu vào prices = [7, 1, 5, 3, 6, 4], hàm sẽ trả về 5, vì bạn có thể mua vào ngày thứ 2 với giá 1 và bán vào ngày thứ 5 với giá 6, lợi nhuận là 6 - 1 = 5.

Với đầu vào prices = [7, 6, 4, 3, 1], hàm sẽ trả về 0, vì không có cơ hội nào để kiếm lợi nhuận (giá chỉ giảm).
