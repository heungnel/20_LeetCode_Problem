## 121. Best Time to Buy and Sell Stock

image.png

```python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        min_price = prices[0]  # Khi to gi mua ti thiu bng gi c phiu u tin
        max_profit = 0  # Khi to li nhun ti a l 0

        # Duyt qua cc mc gi
        for price in prices:
            # Cp nht gi mua ti thiu nu gi hin ti thp hn
            if price < min_price:
                min_price = price
            # Tnh li nhun tim nng v cp nht li nhun ti a nu li nhun ny cao hn
            elif price - min_price > max_profit:
                max_profit = price - min_price
        return max_profit  # Tr v li nhun ti a
```

- min_price c khi to bng gi c phiu u tin, y l gi thp nht m chng ta  thy.
max_profit c khi to l 0, y l li nhun ti a m chng ta s tnh ton.
- S dng vng lp for  duyt tng gi c phiu trong danh sch prices:
Nu gi hin ti (price) thp hn min_price, cp nht min_price.
Tnh li nhun tim nng bng cch ly chnh lch gia gi hin ti v min_price. Nu li nhun tim nng ny ln hn max_profit, cp nht max_profit.
- Sau khi duyt ton b gi c phiu, hm tr v max_profit, tc l li nhun ti a c th t c t giao dch.
