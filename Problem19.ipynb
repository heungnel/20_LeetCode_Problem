## 2078. Two Furthest Houses With Different Colors

image.png

```python
class Solution(object):
    def maxDistance(self, colors):
        """
        :type colors: List[int]
        :rtype: int
        """
        # Khi to t in  lu tr ch s u tin v cui cng ca mi mu
        first_index = {}
        last_index = {}

        # Vng lp qua cc mu  in vo cc ch s u tin v cui cng
        for i in range(len(colors)):
            color = colors[i]
            if color not in first_index:
                first_index[color] = i  # Ghi li ln xut hin u tin
            last_index[color] = i      # Cp nht ln xut hin cui cng

        # Khi to khong cch ti a
        max_dist = 0

        # Tnh khong cch ti a gia cc mu khc nhau
        for color1 in first_index:
            for color2 in last_index:
                if color1 != color2:  # Ch xem xt cc mu khc nhau
                    # Tnh khong cch
                    distance1 = abs(first_index[color1] - last_index[color2])  # Khong cch gia ch s u tin ca color1 v ch s cui cng ca color2
                    distance2 = abs(last_index[color1] - first_index[color2])  # Khong cch gia ch s cui cng ca color1 v ch s u tin ca color2
                    max_dist = max(max_dist, distance1, distance2)  # Cp nht khong cch ti a

        return max_dist  # Tr v khong cch ti a tm c
```

- first_index: T in lu ch s ln xut hin u tin ca mi mu.
last_index: T in lu ch s ln xut hin cui cng ca mi mu.

- S dng vng lp for  duyt qua danh sch colors.
Nu mu cha c trong first_index, lu ch s hin ti (i) lm ch s u tin.
Cp nht ch s cui cng ca mu  vo last_index.

- Khi to bin max_dist  lu khong cch ti a gia cc mu khc nhau.
Duyt qua tng mu trong first_index v last_index  so snh.
Tnh ton khong cch gia ch s u tin ca mu ny v ch s cui cng ca mu kia, cng nh gia ch s cui cng ca mu ny v ch s u tin ca mu kia.

- Cp nht max_dist vi gi tr ln nht gia max_dist, distance1, v distance2.


- Hm tr v khong cch ti a tm c gia cc mu khc nhau.
