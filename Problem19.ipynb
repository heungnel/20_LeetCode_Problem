## 2078. Two Furthest Houses With Different Colors

image.png

```python
class Solution(object):
    def maxDistance(self, colors):
        """
        :type colors: List[int]
        :rtype: int
        """
        # Khởi tạo từ điển để lưu trữ chỉ số đầu tiên và cuối cùng của mỗi màu
        first_index = {}
        last_index = {}

        # Vòng lặp qua các màu để điền vào các chỉ số đầu tiên và cuối cùng
        for i in range(len(colors)):
            color = colors[i]
            if color not in first_index:
                first_index[color] = i  # Ghi lại lần xuất hiện đầu tiên
            last_index[color] = i      # Cập nhật lần xuất hiện cuối cùng

        # Khởi tạo khoảng cách tối đa
        max_dist = 0

        # Tính khoảng cách tối đa giữa các màu khác nhau
        for color1 in first_index:
            for color2 in last_index:
                if color1 != color2:  # Chỉ xem xét các màu khác nhau
                    # Tính khoảng cách
                    distance1 = abs(first_index[color1] - last_index[color2])  # Khoảng cách giữa chỉ số đầu tiên của color1 và chỉ số cuối cùng của color2
                    distance2 = abs(last_index[color1] - first_index[color2])  # Khoảng cách giữa chỉ số cuối cùng của color1 và chỉ số đầu tiên của color2
                    max_dist = max(max_dist, distance1, distance2)  # Cập nhật khoảng cách tối đa

        return max_dist  # Trả về khoảng cách tối đa tìm được
```

- first_index: Từ điển lưu chỉ số lần xuất hiện đầu tiên của mỗi màu.
last_index: Từ điển lưu chỉ số lần xuất hiện cuối cùng của mỗi màu.

- Sử dụng vòng lặp for để duyệt qua danh sách colors.
Nếu màu chưa có trong first_index, lưu chỉ số hiện tại (i) làm chỉ số đầu tiên.
Cập nhật chỉ số cuối cùng của màu đó vào last_index.

- Khởi tạo biến max_dist để lưu khoảng cách tối đa giữa các màu khác nhau.
Duyệt qua từng màu trong first_index và last_index để so sánh.
Tính toán khoảng cách giữa chỉ số đầu tiên của màu này và chỉ số cuối cùng của màu kia, cũng như giữa chỉ số cuối cùng của màu này và chỉ số đầu tiên của màu kia.

- Cập nhật max_dist với giá trị lớn nhất giữa max_dist, distance1, và distance2.


- Hàm trả về khoảng cách tối đa tìm được giữa các màu khác nhau.
