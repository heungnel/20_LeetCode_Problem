## 1394. Find Lucky Integer in an Array

image.png

```python
class Solution(object):
    def findLucky(self, arr):
        """
        :type arr: List[int]
        :rtype: int
        """
        # To mt t in  m tn sut
        frequency = {}

        # m s ln mi s xut hin trong mng
        for num in arr:
            # Nu s  c trong t in, tng s m ln 1
            if num in frequency:
                frequency[num] += 1
            else:
                frequency[num] = 1  # Khi to gi tr 1 nu y l ln u tin chng ta thy s ny

        # Khi to bin  theo di s nguyn may mn ln nht
        largest_lucky = -1  # Bt u vi -1, c ngha l chng ta cha tm thy s nguyn may mn no

        # Kim tra tng s trong t in tn sut
        for num, count in frequency.items():
            if num == count:  # Kim tra xem y c phi l s nguyn may mn khng
                if num > largest_lucky:  # Kim tra xem s nguyn may mn ny c ln hn s ln nht  tm thy khng
                    largest_lucky = num  # Cp nht s nguyn may mn ln nht

        return largest_lucky  # Tr v s nguyn may mn ln nht  tm thy, hoc -1 nu khng c
```

- To mt t in frequency  lu tr s ln xut hin ca mi s trong mng arr.

- S dng vng lp for num in arr:  duyt qua tng s trong mng.
Nu s  c trong t in, tng gi tr m ln 1. Nu cha, khi to gi tr ca n l 1.

- largest_lucky c khi to vi gi tr -1, cho bit rng chng ta cha tm thy s nguyn may mn no.
Kim tra cc s trong t in:

- S dng vng lp for num, count in frequency.items():  kim tra tng s v s ln xut hin ca n.

- Nu s  bng tn sut ca n (if num == count:), th s  c coi l s nguyn may mn.

- Kim tra xem s nguyn may mn ny c ln hn largest_lucky khng. Nu c, cp nht gi tr ca largest_lucky.

- Cui cng, hm tr v gi tr largest_lucky, nu khng tm thy s nguyn may mn no th gi tr ny vn l -1.
