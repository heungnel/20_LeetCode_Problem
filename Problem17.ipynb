## 1394. Find Lucky Integer in an Array

image.png

```python
class Solution(object):
    def findLucky(self, arr):
        """
        :type arr: List[int]
        :rtype: int
        """
        # Tạo một từ điển để đếm tần suất
        frequency = {}

        # Đếm số lần mỗi số xuất hiện trong mảng
        for num in arr:
            # Nếu số đã có trong từ điển, tăng số đếm lên 1
            if num in frequency:
                frequency[num] += 1
            else:
                frequency[num] = 1  # Khởi tạo giá trị 1 nếu đây là lần đầu tiên chúng ta thấy số này

        # Khởi tạo biến để theo dõi số nguyên may mắn lớn nhất
        largest_lucky = -1  # Bắt đầu với -1, có nghĩa là chúng ta chưa tìm thấy số nguyên may mắn nào

        # Kiểm tra từng số trong từ điển tần suất
        for num, count in frequency.items():
            if num == count:  # Kiểm tra xem đây có phải là số nguyên may mắn không
                if num > largest_lucky:  # Kiểm tra xem số nguyên may mắn này có lớn hơn số lớn nhất đã tìm thấy không
                    largest_lucky = num  # Cập nhật số nguyên may mắn lớn nhất

        return largest_lucky  # Trả về số nguyên may mắn lớn nhất đã tìm thấy, hoặc -1 nếu không có
```

- Tạo một từ điển frequency để lưu trữ số lần xuất hiện của mỗi số trong mảng arr.

- Sử dụng vòng lặp for num in arr: để duyệt qua từng số trong mảng.
Nếu số đã có trong từ điển, tăng giá trị đếm lên 1. Nếu chưa, khởi tạo giá trị của nó là 1.

- largest_lucky được khởi tạo với giá trị -1, cho biết rằng chúng ta chưa tìm thấy số nguyên may mắn nào.
Kiểm tra các số trong từ điển:

- Sử dụng vòng lặp for num, count in frequency.items(): để kiểm tra từng số và số lần xuất hiện của nó.

- Nếu số đó bằng tần suất của nó (if num == count:), thì số đó được coi là số nguyên may mắn.

- Kiểm tra xem số nguyên may mắn này có lớn hơn largest_lucky không. Nếu có, cập nhật giá trị của largest_lucky.

- Cuối cùng, hàm trả về giá trị largest_lucky, nếu không tìm thấy số nguyên may mắn nào thì giá trị này vẫn là -1.
