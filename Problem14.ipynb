## 268. Missing Number

image.png

```python
class Solution(object):
    def missingNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)  # Độ dài của mảng nums

        # Tạo một tập hợp từ mảng nums để có thể tìm kiếm nhanh (O(1) thời gian truy xuất)
        num_set = set(nums)

        # Duyệt qua tất cả các số từ 0 đến n + 1
        for i in range(n + 1):
            # Nếu số i không có trong tập hợp num_set, thì đó là số bị thiếu
            if i not in num_set:
                return i  # Trả về số bị thiếu
```

- Tính độ dài của mảng: n = len(nums) tính độ dài của mảng nums, vì ta biết mảng chứa các số từ 0 đến n, nhưng thiếu 1 số duy nhất.
- Tạo tập hợp: num_set = set(nums) chuyển đổi mảng nums thành một tập hợp để có thể kiểm tra sự tồn tại của các phần tử với thời gian O(1).
- Duyệt qua các số từ 0 đến n: for i in range(n + 1) duyệt qua tất cả các số từ 0 đến n. Trong mỗi lần lặp, kiểm tra xem i có tồn tại trong tập hợp num_set hay không bằng cách sử dụng if i not in num_set. Nếu không tồn tại, i chính là số bị thiếu.
- Trả về số bị thiếu: Khi tìm thấy số i không có trong mảng nums, hàm sẽ trả về i.
