## 268. Missing Number

image.png

```python
class Solution(object):
    def missingNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)  #  di ca mng nums

        # To mt tp hp t mng nums  c th tm kim nhanh (O(1) thi gian truy xut)
        num_set = set(nums)

        # Duyt qua tt c cc s t 0 n n + 1
        for i in range(n + 1):
            # Nu s i khng c trong tp hp num_set, th  l s b thiu
            if i not in num_set:
                return i  # Tr v s b thiu
```

- Tnh  di ca mng: n = len(nums) tnh  di ca mng nums, v ta bit mng cha cc s t 0 n n, nhng thiu 1 s duy nht.
- To tp hp: num_set = set(nums) chuyn i mng nums thnh mt tp hp  c th kim tra s tn ti ca cc phn t vi thi gian O(1).
- Duyt qua cc s t 0 n n: for i in range(n + 1) duyt qua tt c cc s t 0 n n. Trong mi ln lp, kim tra xem i c tn ti trong tp hp num_set hay khng bng cch s dng if i not in num_set. Nu khng tn ti, i chnh l s b thiu.
- Tr v s b thiu: Khi tm thy s i khng c trong mng nums, hm s tr v i.
