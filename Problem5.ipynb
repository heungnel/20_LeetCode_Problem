## 88. Merge Sorted Array

image.png

```python
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        """
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        :rtype: None Do not return anything, modify nums1 in-place instead.
        """
        # Bt u hp nht t cui ca c hai mng
        i = m - 1  # Con tr cho phn t cui cng trong nums1
        j = n - 1  # Con tr cho phn t cui cng trong nums2
        k = m + n - 1  # Con tr cho v tr cui cng trong nums1

        # Hp nht cho n khi mt trong cc mng c duyt ht
        while i >= 0 and j >= 0:
            if nums1[i] > nums2[j]:
                nums1[k] = nums1[i]  # Gn phn t ln hn vo v tr k
                i -= 1  # Gim con tr i
            else:
                nums1[k] = nums2[j]  # Gn phn t ln hn vo v tr k
                j -= 1  # Gim con tr j
            k -= 1  # Gim con tr k

        # Nu cn phn t no trong nums2, sao chp chng vo nums1
        while j >= 0:
            nums1[k] = nums2[j]  # Sao chp phn t t nums2 vo nums1
            j -= 1  # Gim con tr j
            k -= 1  # Gim con tr k
```

- i cho phn t cui cng ca nums1, j cho phn t cui cng ca nums2, v k cho v tr cui cng ca nums1.
- S dng vng lp while  so snh cc phn t t cui ca hai mng:
Nu phn t nums1[i] ln hn nums2[j], gn gi tr ny vo v tr k trong nums1, sau  gim i v k.
Ngc li, gn nums2[j] vo nums1[k], sau  gim j v k.
- Nu vn cn phn t trong nums2 sau khi duyt ht nums1, sao chp cc phn t cn li ca nums2 vo nums1.
