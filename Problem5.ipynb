## 88. Merge Sorted Array

image.png

```python
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        """
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        :rtype: None Do not return anything, modify nums1 in-place instead.
        """
        # Bắt đầu hợp nhất từ cuối của cả hai mảng
        i = m - 1  # Con trỏ cho phần tử cuối cùng trong nums1
        j = n - 1  # Con trỏ cho phần tử cuối cùng trong nums2
        k = m + n - 1  # Con trỏ cho vị trí cuối cùng trong nums1

        # Hợp nhất cho đến khi một trong các mảng được duyệt hết
        while i >= 0 and j >= 0:
            if nums1[i] > nums2[j]:
                nums1[k] = nums1[i]  # Gán phần tử lớn hơn vào vị trí k
                i -= 1  # Giảm con trỏ i
            else:
                nums1[k] = nums2[j]  # Gán phần tử lớn hơn vào vị trí k
                j -= 1  # Giảm con trỏ j
            k -= 1  # Giảm con trỏ k

        # Nếu còn phần tử nào trong nums2, sao chép chúng vào nums1
        while j >= 0:
            nums1[k] = nums2[j]  # Sao chép phần tử từ nums2 vào nums1
            j -= 1  # Giảm con trỏ j
            k -= 1  # Giảm con trỏ k
```

- i cho phần tử cuối cùng của nums1, j cho phần tử cuối cùng của nums2, và k cho vị trí cuối cùng của nums1.
- Sử dụng vòng lặp while để so sánh các phần tử từ cuối của hai mảng:
Nếu phần tử nums1[i] lớn hơn nums2[j], gán giá trị này vào vị trí k trong nums1, sau đó giảm i và k.
Ngược lại, gán nums2[j] vào nums1[k], sau đó giảm j và k.
- Nếu vẫn còn phần tử trong nums2 sau khi duyệt hết nums1, sao chép các phần tử còn lại của nums2 vào nums1.

# Ví dụ:
Với đầu vào nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3, mảng cuối cùng sẽ là [1,2,2,3,5,6].

Với đầu vào nums1 = [1], m = 1, nums2 = [], n = 0, mảng cuối cùng sẽ là [1].

Với đầu vào nums1 = [0], m = 0, nums2 = [1], n = 1, mảng cuối cùng sẽ là [1].
