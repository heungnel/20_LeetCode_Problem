## 2928. Distribute Candies Among Children I

image.png

```python
class Solution(object):
    def distributeCandies(self, n, limit):
        """
        :type n: int
        :type limit: int
        :rtype: int
        """
        # Biến này sẽ đếm số cách hợp lệ để phân phối kẹo
        ans = 0

        # Vòng lặp qua tất cả số kẹo có thể cho đứa trẻ đầu tiên
        for x1 in range(0, min(n, limit) + 1):
            # Vòng lặp qua tất cả số kẹo có thể cho đứa trẻ thứ hai
            for x2 in range(0, min(n - x1, limit) + 1):
                # Tính số kẹo còn lại cho đứa trẻ thứ ba
                x3 = n - x1 - x2

                # Kiểm tra xem đứa trẻ thứ ba có nhận được số kẹo hợp lệ không
                if 0 <= x3 <= limit:
                    # Nếu hợp lệ, ta đã tìm ra một cách phân phối kẹo
                    ans += 1

        return ans  # Trả về tổng số cách phân phối hợp lệ
```

- ans: Biến này sẽ lưu tổng số cách phân phối kẹo hợp lệ. Khởi tạo giá trị ban đầu là 0.
- Vòng lặp cho đứa trẻ đầu tiên:

for x1 in range(0, min(n, limit) + 1): Vòng lặp này chạy từ 0 đến số kẹo tối đa mà đứa trẻ đầu tiên có thể nhận, là giá trị nhỏ hơn giữa n (tổng số kẹo) và limit (số kẹo tối đa mỗi đứa trẻ có thể nhận).

- Vòng lặp cho đứa trẻ thứ hai:

for x2 in range(0, min(n - x1, limit) + 1): Tương tự, vòng lặp này chạy cho đứa trẻ thứ hai, nhưng giới hạn số kẹo mà nó có thể nhận sẽ phụ thuộc vào số kẹo còn lại sau khi đã phân phối cho đứa trẻ đầu tiên (n - x1) và cũng không vượt quá limit.

- Tính số kẹo cho đứa trẻ thứ ba:

x3 = n - x1 - x2: Số kẹo còn lại cho đứa trẻ thứ ba sẽ là tổng số kẹo trừ đi số kẹo đã phân phối cho hai đứa trẻ đầu tiên.

- Kiểm tra tính hợp lệ, if 0 <= x3 <= limit: Kiểm tra xem số kẹo x3 cho đứa trẻ thứ ba có hợp lệ hay không (không âm và không vượt quá giới hạn). Nếu hợp lệ, ta đã tìm ra một cách phân phối kẹo.

- Nếu điều kiện hợp lệ, ans sẽ tăng lên 1 để đếm cách phân phối này.
Trả về kết quả:

- Cuối cùng, hàm trả về tổng số cách phân phối hợp lệ mà đã được tính toán.
