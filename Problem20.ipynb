## 2928. Distribute Candies Among Children I

image.png

```python
class Solution(object):
    def distributeCandies(self, n, limit):
        """
        :type n: int
        :type limit: int
        :rtype: int
        """
        # Bin ny s m s cch hp l  phn phi ko
        ans = 0

        # Vng lp qua tt c s ko c th cho a tr u tin
        for x1 in range(0, min(n, limit) + 1):
            # Vng lp qua tt c s ko c th cho a tr th hai
            for x2 in range(0, min(n - x1, limit) + 1):
                # Tnh s ko cn li cho a tr th ba
                x3 = n - x1 - x2

                # Kim tra xem a tr th ba c nhn c s ko hp l khng
                if 0 <= x3 <= limit:
                    # Nu hp l, ta  tm ra mt cch phn phi ko
                    ans += 1

        return ans  # Tr v tng s cch phn phi hp l
```

- ans: Bin ny s lu tng s cch phn phi ko hp l. Khi to gi tr ban u l 0.
- Vng lp cho a tr u tin:

for x1 in range(0, min(n, limit) + 1): Vng lp ny chy t 0 n s ko ti a m a tr u tin c th nhn, l gi tr nh hn gia n (tng s ko) v limit (s ko ti a mi a tr c th nhn).

- Vng lp cho a tr th hai:

for x2 in range(0, min(n - x1, limit) + 1): Tng t, vng lp ny chy cho a tr th hai, nhng gii hn s ko m n c th nhn s ph thuc vo s ko cn li sau khi  phn phi cho a tr u tin (n - x1) v cng khng vt qu limit.

- Tnh s ko cho a tr th ba:

x3 = n - x1 - x2: S ko cn li cho a tr th ba s l tng s ko tr i s ko  phn phi cho hai a tr u tin.

- Kim tra tnh hp l, if 0 <= x3 <= limit: Kim tra xem s ko x3 cho a tr th ba c hp l hay khng (khng m v khng vt qu gii hn). Nu hp l, ta  tm ra mt cch phn phi ko.

- Nu iu kin hp l, ans s tng ln 1  m cch phn phi ny.
Tr v kt qu:

- Cui cng, hm tr v tng s cch phn phi hp l m  c tnh ton.
